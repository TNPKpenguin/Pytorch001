```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sbux.csv", index_col="Date", parse_dates=True)
```

```python
df.head()
```

```python
plt.style.use('ggplot')
df['Volume'].plot(label='CLOSE', title='Star Bucks Stock Volume')
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, 5:6] 
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

mm = MinMaxScaler()
ss = StandardScaler()

X_ss = ss.fit_transform(X)
y_mm = mm.fit_transform(y)
```

```python
# First 200 for training

test_size = 0.2

X_train = X_ss[:int(X_ss.shape[0]*(1-test_size)), :]
X_test = X_ss[int(X_ss.shape[0]*(1-test_size)):, :]

y_train = y_mm[:int(y_mm.shape[0]*(1-test_size)), :]
y_test = y_mm[int(y_mm.shape[0]*(1-test_size)):, :]
```

```python
print("Training Shape", X_train.shape, y_train.shape)
print("Testing Shape", X_test.shape, y_test.shape)
```

```python
import torch
```

```python
X_train_tensor = torch.autograd.Variable(torch.Tensor(X_train))
X_test_tensor = torch.autograd.Variable(torch.Tensor(X_test))

y_train_tensor = torch.autograd.Variable(torch.Tensor(y_train))
y_test_tensor = torch.autograd.Variable(torch.Tensor(y_test))
```

```python
X_train_tensor_final = torch.reshape(X_train_tensor, (X_train_tensor.shape[0], 1, X_train_tensor.shape[1]))
X_test_tensor_final = torch.reshape(X_test_tensor, (X_test_tensor.shape[0], 1, X_test_tensor.shape[1]))
```

```python
print("Training Shape", X_train_tensor_final.shape, y_train_tensor.shape)
print("Testing Shape", X_test_tensor_final.shape, y_test_tensor.shape) 
```

```python
class LSTM1(torch.nn.Module):
    def __init__(self, num_classes, input_size, hidden_size, num_layers, seq_length):
        super(LSTM1, self).__init__()
        self.num_classes = num_classes
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.seq_length = seq_length
    
        self.lstm = torch.nn.LSTM(input_size=input_size,
                                 hidden_size=hidden_size,
                                 num_layers=num_layers,
                                 batch_first=True)
        
        self.fc_1 = torch.nn.Linear(hidden_size, 128)
        self.fc = torch.nn.Linear(128, num_classes)
        
        self.relu = torch.nn.ReLU()
    
    def forward(self, x):
        h_0 = torch.autograd.Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size))
        c_0 = torch.autograd.Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size))
        output, (hn, cn) = self.lstm(x, (h_0, c_0))
        hn = hn.view(-1, self.hidden_size)
        out = self.relu(hn)
        out = self.fc_1(out)
        out = self.relu(out)
        out = self.fc(out)
        return out
```

```python
num_epochs = 3000
learning_rate = 0.001

input_size = 5
hidden_size = 2
num_layers = 1
 
num_classes = 1
```

```python
lstm = LSTM1(num_classes, input_size, hidden_size, num_layers, X_train_tensor_final.shape[1])
```

```python
losses = torch.nn.MSELoss()
optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate)
```

```python
for epoch in range(num_epochs):
    output = lstm.forward(X_train_tensor_final)
    
    loss = losses(output, y_train_tensor)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch%100 == 0:
        print("Epoch: %d, loss: %1.5f" % (epoch, loss.item())) 
```

```python
df_X_ss = ss.transform(df.iloc[:, :-1])
df_y_mm = mm.transform(df.iloc[:, -1:])

df_X_ss = torch.autograd.Variable(torch.Tensor(df_X_ss))
df_y_mm = torch.autograd.Variable(torch.Tensor(df_y_mm))

df_X_ss = torch.reshape(df_X_ss, (df_X_ss.shape[0], 1, df_X_ss.shape[1]))
```

```python
train_predict = lstm(df_X_ss)#forward pass
data_predict = train_predict.data.numpy() #numpy conversion
dataY_plot = df_y_mm.data.numpy()

data_predict = mm.inverse_transform(data_predict) #reverse transformation
dataY_plot = mm.inverse_transform(dataY_plot)
plt.figure(figsize=(10,6)) #plotting
plt.axvline(x=144, c='r', linestyle='--') #size of the training set

plt.plot(dataY_plot, label='Actuall Data') #actual plot
plt.plot(data_predict, label='Predicted Data') #predicted plot
plt.title('Time-Series Prediction')
plt.legend()
plt.show() 
```

```python
class LSTMClassifier(torch.nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = torch.nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = torch.nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out
```

# https://machinelearningmastery.com/lstm-for-time-series-prediction-in-pytorch/

```python
import matplotlib.pyplot as plt 
import pandas as pd
import torch
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Device name : {torch.cuda.get_device_name()}")
```

```python
df = pd.read_csv("airline-passengers.csv")
time_series = df[['Passengers']].values.astype('float32')

plt.plot(time_series)
plt.show()
```

```python
train_size = int(len(time_series)*0.8)
test_size = len(time_series) - train_size
train, test = time_series[:train_size], time_series[train_size:]
```

```python
def create_dataset(dataset, lookback):
    X, y = [], []
    for i in range(len(dataset)-lookback):
        feature = dataset[i:i+lookback]
        target = dataset[i+1:i+lookback+1]
        print(feature, "->", target)
        print("---------")
        X.append(feature)
        y.append(target)
    return torch.tensor(X), torch.tensor(y)
```

```python
lookback = 3
X_train, y_train = create_dataset(train, lookback=lookback)
X_test, y_test = create_dataset(test, lookback=lookback)

X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
class AirModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = torch.nn.LSTM(input_size=1, hidden_size=50, num_layers=1, batch_first=True)
        self.linear = torch.nn.Linear(50, 1)
        
    def forward(self, X):
        x, _ = self.lstm(X)
        x = self.linear(x)
        return x
```

```python
import numpy as np

model = AirModel().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = torch.nn.MSELoss()
data = torch.utils.data.TensorDataset(X_train, y_train)
loader = torch.utils.data.DataLoader(data, shuffle=True, batch_size=8)

n_epochs = 2000
for epoch in range(n_epochs):
    model.train()
    for x, y in loader:
        y_pred = model(x)
#         print(y_pred)
        loss = loss_fn(y_pred, y)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    if epoch % 100 != 0:
        continue
    model.eval()
    with torch.no_grad():
        y_pred = model(X_train)
        train_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_train.cpu()))
        y_pred = model(X_test)
        test_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_test.cpu()))
    print("Epoch %d: train RMSE %.4f, test RMSE %.4f" % (epoch, train_rmse, test_rmse))
```

```python
with torch.no_grad():
    train_plot = np.ones_like(time_series) * np.nan
    y_pred_train = model(X_train).cpu().numpy()  # Move tensor to CPU and convert to numpy array
    y_pred_train = y_pred_train[:, -1, :]
    train_plot[lookback:train_size] = y_pred_train

    test_plot = np.ones_like(time_series) * np.nan
    y_pred_test = model(X_test).cpu().numpy()  # Move tensor to CPU and convert to numpy array
    y_pred_test = y_pred_test[:, -1, :]
    test_plot[train_size + lookback:len(time_series)] = y_pred_test

# plot
plt.plot(time_series, label='Actual Time Series')
plt.plot(train_plot, c='r', label='Train Predictions')
plt.plot(test_plot, c='g', label='Test Predictions')
plt.legend()
plt.show()
```

# ANN --- vs --- LSTM

```python
import torch
import torchvision
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device : {torch.cuda.get_device_name()}")
```

Dataset

```python
df = pd.read_csv("airline-passengers.csv")
time_series = df[['Passengers']].values.astype('float32')

plt.plot(time_series)
plt.show()
```

```python
train_size = int(len(time_series)*0.8)
test_size = len(time_series) - train_size
train, test = time_series[:train_size], time_series[train_size:]
```

```python
def create_data(dataset, lookback):
    X, y = [], []
    for i in range(len(dataset) - lookback):
        data = dataset[i:i+lookback]
        label = dataset[i+1:i+lookback+1]
        X.append(data)
        y.append(label)
    return torch.tensor(X), torch.tensor(y)
```

```python
lookback = 3
X_train, y_train = create_data(train, lookback=lookback)
X_test, y_test = create_data(test, lookback=lookback)

X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
dataset = torch.utils.data.TensorDataset(X_train, y_train)
dataloader = torch.utils.data.DataLoader(dataset, shuffle=True, batch_size=10)
```

Build the model

```python
class ANN(torch.nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = torch.nn.Linear(1, 64)
        self.fc2 = torch.nn.Linear(64, 32)
        self.fc3 = torch.nn.Linear(32, 1)
        self.relu = torch.nn.ReLU()
        
    def forward(self, X):
        x = self.relu(self.fc1(X))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class LSTMClassifier(torch.nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(LSTMClassifier, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = torch.nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc1 = torch.nn.Linear(hidden_size, 32)
        self.fc2 = torch.nn.Linear(32, num_classes)
        self.relu = torch.nn.ReLU()
        
    def forward(self, X):
        x, _ = self.lstm(X)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
ann = ANN().to(device)
lstm = LSTMClassifier(1, 50, 1, 1).to(device)
```

```python
EPOCHS = 1000
LR = 1e-3
loss_fn = torch.nn.MSELoss()
ann_optimizer = torch.optim.Adam(ann.parameters(), lr=LR)
lstm_optimizer = torch.optim.Adam(lstm.parameters(), lr=LR)
```

# Training phase

ann

```python
for epoch in range(EPOCHS):
    running_loss = 0
    ann.train()
    for x, y in dataloader:
        y_hat = ann(x)
        loss = loss_fn(y_hat, y)

        ann_optimizer.zero_grad()
        loss.backward()
        ann_optimizer.step()
        running_loss += loss.item()
    if epoch % 100 != 0:
#         print(f"Epoch {epoch} : loss : {running_loss/len(dataloader)}")
        continue
    ann.eval()
    with torch.no_grad():
        y_pred = ann(X_train)
        train_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_train.cpu()))
        y_pred = ann(X_test)
        test_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_test.cpu()))
    print("Epoch %d: train RMSE %.4f, test RMSE %.4f" % (epoch, train_rmse, test_rmse))
```

lstm

```python
for epoch in range(EPOCHS):
    running_loss = 0
    lstm.train()
    for x, y in dataloader:
        y_hat = lstm(x)
        loss = loss_fn(y_hat, y)

        lstm_optimizer.zero_grad()
        loss.backward()
        lstm_optimizer.step()
        running_loss += loss.item()
    if epoch % 100 != 0:
#         print(f"Epoch {epoch} : loss : {running_loss/len(dataloader)}")
        continue
    lstm.eval()
    with torch.no_grad():
        y_pred = lstm(X_train)
        train_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_train.cpu()))
        y_pred = lstm(X_test)
        test_rmse = np.sqrt(loss_fn(y_pred.cpu(), y_test.cpu()))
    print("Epoch %d: train RMSE %.4f, test RMSE %.4f" % (epoch, train_rmse, test_rmse))
```

```python
ann_pred = ann(X_train)
lstm_pred = lstm(X_train)

ann_pred_test = ann(X_test)
lstm_pred_test = lstm(X_test)
```
